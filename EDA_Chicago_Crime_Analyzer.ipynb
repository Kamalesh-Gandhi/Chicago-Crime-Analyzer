# 1) Univariate Analysis

**Crime Frequency by Type**

```python
# Bar plot for Primary Type
df['Primary Type'].value_counts().plot(kind='barh', figsize=(10, 6))
plt.title('Frequency of Crimes by Type')
plt.xlabel('Number of Crimes')
plt.ylabel('Crime Type')
plt.show()
```

**Crime Severity Distribution**

```python
df['Severity'].value_counts().plot(kind='pie', autopct='%1.1f%%', figsize=(8, 8))
plt.title('Crime Severity Distribution')
plt.ylabel('')
plt.show()
```

# 2) Bivariate Analysis

**Arrest percentage by Primary Type**

```python
arrest_by_type = df.groupby('Primary Type')['Arrest'].mean().sort_values()
arrest_by_type.plot(kind='bar', figsize=(10, 6))
plt.title('Arrest Percentage by Crime Type')
plt.ylabel('Arrest Rate')
plt.xlabel('Crime Type')
plt.show()
```

**Number Crimes by Time of Day**

```python
# Count of crimes by TimeOfDay
df['TimeOfDay'].value_counts().plot(kind='bar', figsize=(8, 6))
plt.title('Crimes by Time of Day')
plt.xlabel('Time of Day')
plt.ylabel('Number of Crimes')
plt.show()
```

# 3) Temporal Analysis

**Crime Trend over the Year**

```python
# Crimes per year
df['Year'].value_counts().sort_index().plot(kind='line', figsize=(10, 6))
plt.title('Crime Trends Over the Years')
plt.xlabel('Year')
plt.ylabel('Number of Crimes')
plt.show()
```

# 4) Spatial Analysis

**Crimes and Arrest by District**

```python
# Calculate total crimes and arrests by district
crime_counts = df['District'].value_counts().sort_index()
arrest_counts = df[df['Arrest']]['District'].value_counts().sort_index()

districts = crime_counts.index  # Get the district names
x = range(len(districts))  # Numeric positions for districts
width = 0.4  # Width of each bar

fig, ax = plt.subplots(figsize=(12, 6))

# Bar for total crimes
ax.bar([pos - width/2 for pos in x], crime_counts, width=width, label='Total Crimes', color='skyblue', alpha=0.7)

# Bar for arrests
ax.bar([pos + width/2 for pos in x], arrest_counts, width=width, label='Arrests', color='orange', alpha=0.7)

# Labels and title
ax.set_title('Crimes and Arrests by Police District')
ax.set_xlabel('District')
ax.set_ylabel('Number of Incidents')
ax.set_xticks(x)
ax.set_xticklabels(districts, rotation=45)
ax.legend()

# Show the plot
plt.tight_layout()
plt.show()
```

**Disparity in Crime and Arrest Numbers**: The graph highlights a significant gap between total crimes and arrests in every district, indicating that a large proportion of crimes do not result in arrests.

**High-Crime Districts**: Districts like 2, 10, and 25 have the highest number of crimes, yet their arrest counts are not proportional, suggesting potential inefficiencies or challenges in enforcement.

**Low-Crime Districts**: Districts such as 14, 18, and 31 have both low crime rates and low arrest numbers, which could indicate better policing or fewer reported incidents in these areas.

# 5) Severity Analysis

**Crime Severity by primary Type**

```python
# Group by Primary Type and Severity, and calculate the count
severity_by_type = df.groupby('Primary Type')['Severity'].value_counts().unstack()

# Calculate the total crimes for each Primary Type and sort
severity_by_type['Total'] = severity_by_type.sum(axis=1)
severity_by_type = severity_by_type.sort_values(by='Total', ascending=False).drop(columns='Total')

# Define severity-related colors
severity_colors = {
    'Severe': 'red',
    'Moderate': 'orange',
    'Low': 'green'
}

# Plot a horizontal stacked bar chart
severity_by_type.plot(
    kind='barh',
    stacked=True,
    figsize=(12, 8),
    color=[severity_colors.get(severity, 'gray') for severity in severity_by_type.columns]
)

# Add labels and title
plt.title('Crime Severity by Primary Type (Sorted by Total Crimes)')
plt.xlabel('Number of Crimes')
plt.ylabel('Crime Type')
plt.legend(title='Severity', bbox_to_anchor=(1.05, 1), loc='upper left')  # Adjust legend position
plt.tight_layout()  # Adjust layout to fit everything nicely
plt.show()
```

**Crime Severity By District**

```python
# Group by Primary Type and Severity, and calculate the count
severity_by_district = df.groupby('District')['Severity'].value_counts().unstack()

# Calculate the total crimes for each Primary Type and sort
severity_by_district['Total'] = severity_by_district.sum(axis=1)
severity_by_district = severity_by_district.sort_values(by='Total', ascending=False).drop(columns='Total')

# Define severity-related colors
severity_colors = {
    'Severe': 'red',
    'Moderate': 'orange',
    'Low': 'green'
}

# Plot a horizontal stacked bar chart
severity_by_district.plot(
    kind='barh',
    stacked=True,
    figsize=(12, 8),
    color=[severity_colors.get(severity, 'gray') for severity in severity_by_district.columns]
)

# Add labels and title
plt.title('Crime Severity by District')
plt.xlabel('Number of Crimes')
plt.ylabel('District')
plt.legend(title='Severity', bbox_to_anchor=(1.05, 1), loc='upper left')  # Adjust legend position
plt.show()
```

# 6) Location Analysis

**Blocks with Highest Crime rates**

```python
# Top locations with the highest crime rates
top_blocks = df['Block'].value_counts().head(10)
top_blocks.plot(kind='bar', figsize=(12, 6), color='skyblue')
plt.title('Top 10 Crime Hotspots by Block')
plt.xlabel('Block')
plt.ylabel('Number of Crimes')
plt.xticks(rotation=45)
plt.show()
```

**Crime trends by Arrest vs. No Arrest**

```python
arrest_trends = df.groupby(['Year', 'Arrest']).size().unstack()
arrest_trends.plot(kind='line', figsize=(10, 6), marker='o')
plt.title('Crime Trends: Arrest vs. No Arrest')
plt.xlabel('Year')
plt.ylabel('Number of Crimes')
plt.legend(['No Arrest', 'Arrest'])
plt.grid(True)
plt.show()
```

**Crime Trend by overall Hour of the Day**

```python
df.groupby('Hour').size().plot(kind='line', figsize=(10, 6), marker='o', color='brown')
plt.title('Crimes by Hour of the Day')
plt.xlabel('Hour')
plt.ylabel('Number of Crimes')
plt.grid(True)
plt.show()
```

**Crime count vs Location Description**

```python
# Crime count vs Indoor and Outdoor locations
location_types = df.groupby('Location Description').size().sort_values(ascending=False).head(10)
location_types.plot(kind='bar', figsize=(12, 6), color='skyblue')
plt.title('Crime Count by Location Description')
plt.xlabel('Location Type')
plt.ylabel('Number of Crimes')
plt.xticks(rotation=45)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Calculate arrest rates for domestic and non-domestic crimes
domestic_arrests = df[df['Domestic'] == True]['Arrest'].mean()
non_domestic_arrests = df[df['Domestic'] == False]['Arrest'].mean()

# Data for plotting
categories = ['Domestic', 'Non-Domestic']
arrest_rates = [domestic_arrests, non_domestic_arrests]

# Plotting
plt.figure(figsize=(8, 6))
plt.bar(categories, arrest_rates, color=['skyblue', 'orange'])
plt.title("Arrest Rates for Domestic vs Non-Domestic Crimes")
plt.xlabel("Crime Type")
plt.ylabel("Arrest Rate")
plt.ylim(0, 1)  # Arrest rates are between 0 and 1
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
