Handle Spelling and Repeated values in the Each Columns

1) Primary Type Column

```python
#Unique Values and their respective Counts
df['Primary Type'].value_counts()
```

```python
#Handling Some the values in the Primary Type Column
df['Primary Type'] = df['Primary Type'].replace({'CRIM SEXUAL ASSAULT':'CRIMINAL SEXUAL ASSAULT',
                                                 'OTHER NARCOTIC VIOLATION':'NARCOTICS',
                                                 'PUBLIC INDECENCY': 'SEXUAL MISCONDUCT',
                                                 'OBSCENITY': 'SEXUAL MISCONDUCT',
                                                 'RITUALISM': 'RARE OFFENSE',
                                                 'NON-CRIMINAL':'RARE OFFENSE',
                                                 'DOMESTIC VIOLENCE':'RARE OFFENSE'})
```

```python
#Checking Unique Values and their respective Counts after handling the values
df['Primary Type'].value_counts()
```

2)Block Column

```python
#Find the Count of Unique Values in the Block Column
df['Block'].nunique()
```

```python
# The Reason for Converting  all the values in the 'Block' column to uppercasem is to ensure that the values are uniform
df['Block'] = df['Block'].str.upper()
```

3) Description Column

```python
# Unique Values and their respective Counts
df['Description'].value_counts()
```

```python
# Normalization Dictionary
normalization_map = {
    "ARMED: HANDGUN": "ARMED HANDGUN",
    "ARMED - HANDGUN": "ARMED HANDGUN",
    "POSS: CANNABIS 30GMS OR LESS": "POSSESSION: CANNABIS 30GMS OR LESS",
    "POSS: CANNABIS MORE THAN 30GMS": "POSSESSION: CANNABIS MORE THAN 30GMS",
    "FINANCIAL IDENTITY THEFT OVER $ 300": "FINANCIAL IDENTITY THEFT",
    "FINANCIAL IDENTITY THEFT: OVER $300": "FINANCIAL IDENTITY THEFT",
    "OVER $500": "THEFT OVER $500",
    "$500 AND UNDER": "THEFT $500 AND UNDER"
}

# Apply Normalization
df['Description'] = df['Description'].replace(normalization_map)
```
